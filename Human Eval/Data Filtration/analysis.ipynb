```python
import pandas as pd
import os
from sklearn.metrics import classification_report, cohen_kappa_score
```

```python
if not os.path.exists("Samples/filtered_USPolitics.xlsx"):
    df_MNLI = pd.read_csv("../../Dataset Processing/Outputs/notes_MNLI.tsv", sep="\t", low_memory=False).sample(n=1000, random_state=42).reset_index()
    df_Keywords = pd.read_csv("../../Dataset Processing/Outputs/notes_Keywords.tsv", sep="\t", low_memory=False).sample(n=1000, random_state=42).reset_index()
    df = pd.read_csv("../../Dataset Processing/Outputs/notes_MNLI_Keywords.tsv", sep="\t", low_memory=False).sample(n=1000, random_state=42).reset_index()

    df_human_eval = df[["summary", "isUSPolitics"]].copy()
    df_human_eval["isUSPolitics_MNLI"] = df_MNLI["isUSPolitics"].values
    df_human_eval["isUSPolitics_Keywords"] = df_Keywords["isUSPolitics"].values
    df_human_eval["human"] = ""

    df_human_eval.to_excel("Samples/filtered_USPolitics.xlsx", index=False)
else:
    print("File Samples/filtered_USPolitics.xlsx already exists. Annotate it manually (under human column), if not done already.")
```

```python
df = pd.read_excel('Samples/filtered_USPolitics_annotated.xlsx')

# df.dropna(inplace=True)

df.human = df.human.map({0: 'Other', 1: 'United States Politics'})
```

```python
y_true = df['human']
y_pred = df['isUSPolitics']
# Print classification report
print("Classification Report:")
print(classification_report(y_true, y_pred, digits=3))

# Compute and print Cohen's Kappa
kappa = cohen_kappa_score(y_true, y_pred)
print(f"Cohen's Kappa: {kappa:.3f}")
```

```python
y_true = df['human']
y_pred = df['isUSPolitics_MNLI']
# Print classification report
print("Classification Report:")
print(classification_report(y_true, y_pred, digits=3))

# Compute and print Cohen's Kappa
kappa = cohen_kappa_score(y_true, y_pred)
print(f"Cohen's Kappa: {kappa:.3f}")
```

```python
y_true = df['human']
y_pred = df['isUSPolitics_Keywords']
# Print classification report
print("Classification Report:")
print(classification_report(y_true, y_pred, digits=3))

# Compute and print Cohen's Kappa
kappa = cohen_kappa_score(y_true, y_pred)
print(f"Cohen's Kappa: {kappa:.3f}")
```
